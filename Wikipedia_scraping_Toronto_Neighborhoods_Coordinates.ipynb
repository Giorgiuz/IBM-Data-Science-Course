```python
# import the library we use to open URLs
import urllib.request
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
page = urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
all_tables = soup.find_all('table')
```

```python
right_table=soup.find('table', class_='wikitable sortable')
```

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd

df=pd.DataFrame(A,columns=['Postal Code'])
df['Bourough']=B
df['Neighborhood']=C

df
```

```python
df = df.replace('\n','', regex=True)
df.head()
```

```python
df_2 = df[df['Bourough'] != 'Not assigned'].reset_index(drop=True)
df_2.head()
```

```python
df_2['Neighborhood'].replace('',df_2['Bourough'])
df_2['Neighborhood'].replace('Not assigned',df_2['Bourough'])
df_2.head(12)
```

```python
print('Database shape is: ',df_2.shape)
```

```python
url="https://cocl.us/Geospatial_data"
coordinates = pd.read_csv(url)
coordinates.head()
```

```python
df_3 = pd.merge(left=df_2, right=coordinates, how='left', left_on='Postal Code', right_on='Postal Code')
df_3.head(12)
```
